ATIVIDADE 2
- Qual é a média, a moda e a mediana de horas de sono para cada uma das profissões? [‘mean’, np.median, pd.Series.mod]

```python
grupo_profissao = df.groupby('Profissão')
```

```python
media = grupo_profissao['Duração do sono'].mean()
mediana = grupo_profissao['Duração do sono'].median()
moda = grupo_profissao['Duração do sono'].apply(lambda x: x.mode()[0])
```

```python
resultados = pd.DataFrame({
    'Média': media,
    'Mediana': mediana,
    'Moda': moda
})
print(resultados)
```

Atividade *3*
- Das pessoas que atuam com engenharia de software qual a porcentagem de obesos?

```python
eng_software = df[df['Profissão'] == 'Eng. de Software']
```

```python
obesos_eng_software = eng_software[eng_software['Categoria IMC'] == 'Obesidade']
```

```python
porcentagem_obesos = (len(obesos_eng_software) / len(eng_software)) * 100
print(f"A porcentagem de engenheiros de software obesos é: {porcentagem_obesos:.2f}%")
```

```python
df[['Pressão Sistólica', 'Pressão Diastólica']] = df['Pressão sanguínea'].str.split('/', expand=True)

# Converter as colunas 'Pressão Sistólica' e 'Pressão Diastólica' em tipos numéricos
df['Pressão Sistólica'] = pd.to_numeric(df['Pressão Sistólica'])
df['Pressão Diastólica'] = pd.to_numeric(df['Pressão Diastólica'])

# Calcular a média da pressão sanguínea para cada gênero
media_pressao_sanguinea_por_genero = df.groupby('Gênero').agg({'Pressão Sistólica': 'mean', 'Pressão Diastólica': 'mean'})
```

Atividade 9

É predominante entre os participantes dormir 8 horas por dia
(considere usar Moda como medida)?

```python
moda_horas_sono = df['Duração do sono'].mode()[0]
if moda_horas_sono == 8:
    print("Dormir 8 horas por dia é predominante entre os participantes.")
else:
    print("Dormir 8 horas por dia não é predominante entre os participantes.")
```

Atividade 10

 Pessoas com frequências cardíacas acima de 70 dão mais passos que
pessoas com frequência cardíaca menor ou igual a 70? (Use a média)

```python
media_passos_acima_70 = df[df['Frequência cardíaca'] > 70]['Passos diários'].mean()

# Calcular a média de passos para pessoas com frequências cardíacas menor ou igual a 70
media_passos_70_ou_menos = df[df['Frequência cardíaca'] <= 70]['Passos diários'].mean()

if media_passos_acima_70 > media_passos_70_ou_menos:
    print("Pessoas com frequências cardíacas acima de 70 dão mais passos do que pessoas com frequência cardíaca menor ou igual a 70.")
elif media_passos_acima_70 < media_passos_70_ou_menos:
    print("Pessoas com frequências cardíacas acima de 70 dão menos passos do que pessoas com frequência cardíaca menor ou igual a 70.")
else:
    print("Pessoas com frequências cardíacas acima de 70 dão o mesmo número de passos do que pessoas com frequência cardíaca menor ou igual a 70.")
```
